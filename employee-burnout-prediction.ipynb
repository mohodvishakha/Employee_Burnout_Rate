```python
Employee_burnout_filepath = '/kaggle/input/are-your-employees-burning-out/train.csv'

Employee_burnout_data = pd.read_csv(Employee_burnout_filepath)

Employee_burnout_data.head()
```

```python
Employee_burnout_data.columns= Employee_burnout_data.columns.str.replace(" ", "_")     # Replace columns names with underscore

Employee_burnout_data = Employee_burnout_data.dropna(axis=0)
```

```python
Employee_burnout_data.columns   #to get the column names in data set
```

```python
y = Employee_burnout_data.Burn_Rate #selecting the prediction target
#selecting features
Employee_features = [ 'Designation', 'Resource_Allocation',
       'Mental_Fatigue_Score']                   
X=Employee_burnout_data[Employee_features]
X.describe()
```

```python
#create a model,fit a model, predict values
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
Emp_burnout_model = DecisionTreeRegressor()
Emp_burnout_model.fit(train_X,train_y)
val_prediction = Emp_burnout_model.predict(val_X)
print(r2_score(val_y, val_prediction))

```
